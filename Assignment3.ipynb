```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Preprocessing libries
from sklearn.preprocessing import StandardScaler # for normalization
```

```python
#splitiing the dataset
from sklearn.model_selection import train_test_split
```

```python
# the metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
data= pd.read_csv("/content/petrol_consumption.csv")
```

```python
# Revealing the dataset
print(data)
```

```python
data.columns
```

```python
#splititng the datset
X =data.drop(['Petrol_Consumption'], axis = 1)
y = data['Petrol_Consumption']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.3, random_state= 42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#normaliztion
scaler =StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#model buliding
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred = linear_model.predict(X_test)
```

```python
mse = mean_absolute_error(y_test, y_pred)
mse
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
#model buliding
KNeighbors_model = KNeighborsRegressor()
KNeighbors_model.fit(X_train, y_train)
y_pred = KNeighbors_model.predict(X_test)
```

```python
mse = mean_absolute_error(y_test, y_pred)
mse
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
#model buliding
DecisionTree_model = DecisionTreeRegressor()
DecisionTree_model.fit(X_train, y_train)
y_pred = DecisionTree_model.predict(X_test)
```

```python
mse = mean_absolute_error(y_test, y_pred)
mse
```
